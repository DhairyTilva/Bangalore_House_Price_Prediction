```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_csv('bengaluru_house_prices.csv')
df.head()
```

```python
df1 = df.drop(['area_type','availability','balcony','society'],axis=1)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2 = df1.dropna()
df2.isna().sum()
```

```python
df2.shape
```

```python
df2['size'].unique()
```

```python
df2['bhk'] = df2['size'].apply(lambda x : int(x.split(' ')[0]))
```

```python
df2.head()
```

```python
df2['bhk'].unique()
```

```python
df2['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2[~df2['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return ((float(tokens[0]) + float(tokens[1])) / 2)
    try:
        return float(x)
    except:
        return None
```

```python
df3 = df2.copy()
```

```python
df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft_to_num)
df3.head(10)
```

```python
df3.dropna(inplace=True)
```

```python
df3.isna().sum()
```

```python
df4 = df3.copy()
```

```python
df4['price_per_sqft'] = df4['price']*100000 / df4['total_sqft']
df4.head()
```

```python
df4.location.nunique()
```

```python
df4['location'] = df4['location'].apply(lambda x : x.strip())
df4.head()
```

```python
location_stats = df4.groupby('location')['location'].count().sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats <= 10])
```

```python
location_stats_less_than_10 = location_stats[location_stats <= 10]
location_stats_less_than_10
```

```python
df4['location'] = df4['location'].apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
```

```python
df4['location'].nunique()
```

```python
df4[df4['total_sqft']/df4['bhk'] < 300].head()
```

```python
df4.shape
```

```python
df5 = df4[~(df4['total_sqft']/df4['bhk'] < 300)]
df5.shape
```

```python
df5['price_per_sqft'].describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        mean = np.mean(subdf['price_per_sqft'])
        std = np.std(subdf['price_per_sqft'])
        reduced_df = subdf[(subdf['price_per_sqft'] > (mean - std)) & (subdf['price_per_sqft'] <= (mean + std))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

df6 = remove_pps_outliers(df5)
df6.shape
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df['price_per_sqft']<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

df7 = remove_bhk_outliers(df6)
df7.shape
```

```python
plt.hist(df7['price_per_sqft'],rwidth=0.8)
plt.xlabel('price per sqft')
plt.xlabel('count')
```

```python
df8 = df7[df7['bath'] < df7['bhk']+2]
df8.shape
```

```python
df9 = df8.drop(['size','price_per_sqft'],axis=1)
df9.head()
```

```python
dummies = pd.get_dummies(df9['location'],dtype=np.int32)
dummies
```

```python
df10 = pd.concat([df9.drop('location',axis=1),dummies.drop('other',axis=1)],axis=1)
df10.head()
```

```python
df10.shape
```

```python
X = df10.drop('price',axis=1)
y = df10['price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 5, test_size=0.2,random_state=10)

cross_val_score(LinearRegression(),X,y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'n_jobs': [1,2]
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr.predict([x])[0]
```
